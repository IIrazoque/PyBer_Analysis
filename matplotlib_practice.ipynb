```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
#set x-axis to a list of strings
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#set y-axis to a list of floats
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
#creating a graph using Object Oriented Interface Method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#another way to create graph
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#another way
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#create a plot and add label for the legend 
plt.plot(x_axis, y_axis, label='Boston', color="blue", marker="*", linewidth=2)

#labels for axis 
plt.xlabel("Date")
plt.ylabel("Fare ($)")

#set y limit
plt.ylim(0, 45)

#title
plt.title("PyBer Fare by Month")
plt.grid()

#add legend
plt.legend()
```

```python
#SKILL DRILL create a plot and add label for the legend 
plt.plot(x_axis, y_axis, label='Boston', color="green", marker="d", linewidth=2)

#labels for axis 
plt.xlabel("Date")
plt.ylabel("Fare ($)")

#set y limit
plt.ylim(0, 45)

#title
plt.title("PyBer Fare by Month")
plt.grid()

#add legend
plt.legend()
```

```python
#create bar chart 
plt.bar(x_axis, y_axis, label='Boston', color="green", linewidth=2)

#labels for axis 
plt.xlabel("Date")
plt.ylabel("Fare ($)")

#set y limit
plt.ylim(0, 45)

#title
plt.title("PyBer Fare by Month")


#add legend
plt.legend()
```

```python
#create a horizontal bar chart
plt.barh(x_axis, y_axis)

#invert the y axis 
plt.gca().invert_yaxis()
```

```python
#SKILL DRILL create bar chart 
plt.barh(x_axis, y_axis, label='Boston', color="magenta", linewidth=2)
plt.gca().invert_yaxis()
#labels for axis 
plt.xlabel("Date")
plt.ylabel("Fare ($)")

#title
plt.title("PyBer Fare by Month")


#add legend
plt.legend()
```

```python
#creat bar graph using OOI 
fig, ax = plt.subplots()

ax.bar(x_axis, y_axis, color="k", label='New York')
```

```python
#create horizontal bar graph using OOI
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
#SKILL DRILL create horizontal bar graph using OOI
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.legend()
plt.gca().invert_yaxis
```

```python
#SCATTER PLOTS
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
#SKILL DRILL 
y_axis_larger = []

for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger, label='Chicago', color="red")
plt.gca().invert_yaxis


#labels for axis 
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
#SKILL DRILL 
#y_axis_larger = []

#for data in y_axis:
#y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis], label='Chicago', color="red")
plt.gca().invert_yaxis


#labels for axis 
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
#scatter plot using OOI
fig, ax=plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#create bubble chart using OOI
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
#SKILL DRILL 
fig, ax=plt.subplots()
ax.scatter(x_axis,
           y_axis, 
           c="lightskyblue",
           edgecolors="black",
           alpha=0.2,
           linewidth=2,
           s= [i*5 for i in y_axis],
           label="Boston",)

plt.gca().invert_yaxis

ax.set_xlabel("Fare ($)")
ax.set_ylabel("Months")
ax.set_title("PyBer Fare by Month")
ax.legend()
```

```python
#pie charts
plt.subplots(figsize=(8,8))
explode = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, labels=x_axis, explode=explode, autopct='%.1f%%')
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.pie(y_axis, labels=x_axis, explode=explode, colors=colors, autopct='%.1f%%')
plt.show()
```

```python
#Pie charts using OOI
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#SKILL DRILL 

fig, ax = plt.subplots(figsize=(8, 8))
explode_values=(0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors=["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis, 
       labels=x_axis,
       shadow=True,
       colors=colors,
       explode=explode_values,
       counterclock=False,
       startangle=90,
       autopct='%.1f%%')

plt.gca().invert_yaxis

plt.show()
```
